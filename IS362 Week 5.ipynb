QUESTION 3: On February 12th, 2013, which New York area airport had the windiest weather?

```python
import pandas as pd
```

Load weather.csv

```python
weather = pd.read_csv("weather.csv")
weather.head()
```

Filter for Februart 12, 2013

```python
weather_02122013 = weather[
    (weather['year'] == 2013) &
    (weather['month'] == 2) &
    (weather['day'] == 12)
]
```

Determine highest wind speed and removing inaccurate wind speeds.

```python
weather_clean = weather_02122013[weather_02122013['wind_speed'] < 200]
weather_clean.groupby('origin')['wind_speed'].max()
```

After filtering the dataset for February 12, 2013 and removing unrealistic outlier values, I compared the recorded wind speeds for the three New York area airports. LaGuardia Airport (LGA) had the highest wind speed at approximately 23.02 mph, making it the windiest airport on that date.
